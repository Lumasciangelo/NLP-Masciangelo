```python
import requests
from bs4 import BeautifulSoup
import lxml
from PIL import Image
import pytesseract
from urllib.parse import urljoin
from io import BytesIO
import urllib3
import urllib
```

```python
# Desactivar las advertencias de solicitudes inseguras
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

url = 'https://bitacora.yvera.tur.ar/posts/2023-08-17-ods-turismo/'

response = requests.get(url, verify=False)
soup = BeautifulSoup(response.text, 'html.parser')

img_tabla = soup.find('img')
url_imagen = urljoin(url, img_tabla['src'])


imagen_respuesta = requests.get(url_imagen, verify=False)
imagen = Image.open(BytesIO(imagen_respuesta.content))

imagen.show()

imagen.save('imagen.png')
```

```python
import PyPDF2

# Abre el archivo en modo binario de lectura ('rb')
with open('William S. Cleveland-Visualizing data-Hobart Press (1993).pdf', 'rb') as archivo:
    # Crea un objeto PdfFileReader
    lector = PyPDF2.PdfReader(archivo)

    # Inicializa una cadena vaca para almacenar el texto
    texto = ''

    # Solo voy a obtener la pagina 8 para que no sea tan largo
    # Obtiene la pgina 8 del PDF (indico 8 porque empieza en 0)
    pagina = lector.pages[7]

    # Extrae el texto de la pgina y lo aade a la cadena de texto
    texto += pagina.extract_text ()

# Imprime el texto extrado
print(texto)
```

```python
# Ruta al archivo PDF
archivo_pdf = 'dokumen.tips_tecnicas-muestreo-cochran-parte1pdf.pdf'

# Convierte el PDF en una lista de imgenes
# Esto me va a guardar slo la pagina 18 para que no sea tan largo
imagen = convert_from_path(archivo_pdf, first_page=18, last_page=18)[0]

imagen.save('pagina18.png', 'PNG')

# # Abre el archivo en modo binario de lectura ('rb')
# with open(archivo_pdf, 'rb') as archivo:
#     # Crea un objeto PdfFileReader
#     lector = PyPDF2.PdfReader(archivo)
```

```python
import numpy as np
import easyocr

# Como el PDF tiene dos pginas del libro por pgina, tengo que recuperar la primer mitad de la pgina y leer eso
# Crea un lector OCR en espaol
lector = easyocr.Reader(['es'])

# Ruta a la imagen
archivo_imagen = 'pagina18.png'

# Abre la imagen
imagen = Image.open(archivo_imagen)

# Calcula las coordenadas de la regin de inters (primera mitad de la pgina 18)
ancho, alto = imagen.size
coord_x1 = 0
coord_y1 = 0
coord_x2 = ancho // 2
coord_y2 = alto

# Recorta la imagen para obtener la primera mitad de la pgina
region_interes = imagen.crop((coord_x1, coord_y1, coord_x2, coord_y2))

# Convierte la imagen recortada a un arreglo numpy
imagen_np = np.array(region_interes)

# Usa EasyOCR para convertir la regin de la imagen en texto
resultado = lector.readtext(imagen_np, detail=0)

# Imprime el texto extrado
print("Texto extrado de la primera mitad de la pgina 18:")
print(resultado)
```
